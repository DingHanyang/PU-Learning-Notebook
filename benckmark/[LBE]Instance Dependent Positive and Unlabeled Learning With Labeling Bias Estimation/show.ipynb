```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 strategy1 函数
def strategy1(x, theta_lgt, k):
    eta_x = (1 + np.exp(-np.dot(theta_lgt.T, x)))**(-1/k)
    return eta_x

# 定义参数
theta_lgt = np.array([1, 1])  # 示例参数，可以根据需要调整
k = 1

# 创建二维网格
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
X = np.array([X1.ravel(), X2.ravel()])

# 计算函数值
eta_values = strategy1(X, theta_lgt, k).reshape(X1.shape)

# 绘制图像
plt.figure(figsize=(6, 6))
plt.contourf(X1, X2, eta_values, levels=50, cmap='viridis')
plt.colorbar()
plt.title(r'True $\eta(\mathbf{x})$')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.show()
```
